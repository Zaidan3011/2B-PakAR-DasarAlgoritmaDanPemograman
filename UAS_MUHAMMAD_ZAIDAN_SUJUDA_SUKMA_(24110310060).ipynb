```python
import json

class InventoryItem:
    def __init__(self, name, stock, price):
        self.name = name
        self.stock = stock
        self.price = price

    def to_dict(self):
        return {
            'name': self.name,
            'stock': self.stock,
            'price': self.price
        }

class Inventory:
    def __init__(self):
        self.items = []

    def add_item(self, item):
        self.items.append(item)

    def display_items(self):
        for index, item in enumerate(self.items):
            print(f"{index + 1}. {item.name} - Stok: {item.stock}, Harga: {item.price}")

    def edit_item(self, index, name, stock, price):
        if 0 <= index < len(self.items):
            self.items[index].name = name
            self.items[index].stock = stock
            self.items[index].price = price

    def delete_item(self, index):
        if 0 <= index < len(self.items):
            del self.items[index]

    def search_item(self, name):
        return [item for item in self.items if item.name.lower() == name.lower()]

    def total_inventory_value(self):
        return sum(item.stock * item.price for item in self.items)

    def export_to_file(self, filename):
        with open(filename, 'w') as f:
            json.dump([item.to_dict() for item in self.items], f)

def main():
    inventory = Inventory()

    while True:
        print("\nMenu:")
        print("1. Tambah Barang")
        print("2. Tampilkan Daftar Barang")
        print("3. Edit Barang")
        print("4. Hapus Barang")
        print("5. Cari Barang")
        print("6. Laporan Ringkas")
        print("7. Export ke File")
        print("8. Keluar")

        choice = input("Pilih opsi: ")

        if choice == '1':
            name = input("Nama Barang: ")
            stock = int(input("Stok: "))
            price = float(input("Harga: "))
            inventory.add_item(InventoryItem(name, stock, price))
        elif choice == '2':
            inventory.display_items()
        elif choice == '3':
            index = int(input("Nomor Barang yang ingin diedit: ")) - 1
            name = input("Nama Baru: ")
            stock = int(input("Stok Baru: "))
            price = float(input("Harga Baru: "))
            inventory.edit_item(index, name, stock, price)
        elif choice == '4':
            index = int(input("Nomor Barang yang ingin dihapus: ")) - 1
            inventory.delete_item(index)
        elif choice == '5':
            name = input("Nama Barang yang dicari: ")
            results = inventory.search_item(name)
            for item in results:
                print(f"Ditemukan: {item.name} - Stok: {item.stock}, Harga: {item.price}")
        elif choice == '6':
            total_items = len(inventory.items)
            total_value = inventory.total_inventory_value()
            print(f"Jumlah Total Barang: {total_items}, Total Nilai Inventori: {total_value}")
        elif choice == '7':
            filename = input("Nama file untuk export (misal: inventory.json): ")
            inventory.export_to_file(filename)
            print("Data berhasil diekspor.")
        elif choice == '8':
            break
        else:
            print("Pilihan tidak valid.")

if __name__ == "__main__":
    main()
```
